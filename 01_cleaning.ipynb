# Data Cleaning

There are a few issues with the data such as missing CustomerID's, and a wide variety of refunds (Invoices which start with 'C'), and other rows which have negative quantities or prices.

Invoices that start with C are removed because they distort revenue and purchase frequency statistics. In general, negative quantity and prices are removed for this reason as well.

Data which is missing a CustomerID is kept unless it also has a negative quantity or price because it still contributes to overall sales data even if it cannot be attributed to a specific customer. 


```python
import pandas as pd
import datetime as dt
```

```python
df1 = pd.read_excel('data/online_retail_II.xlsx', sheet_name='Year 2009-2010')
df2 = pd.read_excel('data/online_retail_II.xlsx', sheet_name='Year 2010-2011')
df = pd.concat([df1, df2])
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df = df[df['Quantity'] > 0]
```

```python
df = df[df['Price'] > 0]
```

```python
df['Purchase Month'] = df['InvoiceDate'].dt.to_period('M')
df['Revenue'] = df['Price'] * df['Quantity']
```

```python
df.to_csv('data/online_retail_II_clean.csv', index=False)
```
