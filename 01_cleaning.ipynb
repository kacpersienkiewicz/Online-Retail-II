# Data Cleaning

Data Cleaning was handled in Excel and included two things:
* Removal of rows with negative quantity, price, or was prepended with a letter which meant that it was an administrative cost or return. Since I wanted to look at revenue patterns, I wanted to remove any returns or administrative costs that are more relevant to profit or expense analyses.
* Feature engineering two new columns: Payment Month (allows for grouping by purchase month rather than just the day) and revenue (provides the total revenue for the order).


```python
import pandas as pd
```

```python
df1 = pd.read_excel('data/online_retail_II_clean.xlsx', sheet_name='Year 2009-2010_clean')
df2 = pd.read_excel('data/online_retail_II_clean.xlsx', sheet_name='Year 2010-2011_clean')
df = pd.concat([df1, df2])
```
